```python
import pandas as pd
import numpy as np
import pickle
from pymongo import MongoClient
```

```python
client = MongoClient("mongodb://localhost:27017/")
db = client.crypto
collection = db.crypto_collection
```

```python
ticks = pd.DataFrame.from_dict(collection.find({'type':'ticker', 'product_id':'BTC-USD'}))
```

```python
ticks.head()
```

```python
ticks.info()
```

```python
ticks['dt'] = pd.to_datetime(ticks.time)
```

```python
ticks['dt_min'] = ticks.dt.dt.floor('60s')
```

```python
ticks.groupby('dt_min').price.agg(['max','min'])
```
